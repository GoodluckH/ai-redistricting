```python
import numpy as np
import pickle
import os
import re
pattern = r"sbl(\d+)_scorenum_opportunity_dists_([0-9.]+).(npy|p)"

for file_name in os.listdir("../data/"):
    if file_name.endswith(".npy") or file_name.endswith(".p"):
        match = re.search(pattern, file_name)

        if match:
            sbl = match.group(1)
            threshold = match.group(2)

            print(f"\n\nShort Burst Length: {sbl}, Threshold: {threshold}")
            if file_name.endswith(".npy"):
                res = np.load(f"../data/{file_name}")
                print(f"Max Score: {np.max(res)}")

            else:
                with open("../data/OH_dists15_BVAPopt_2.0%_20000_sbl15_scorenum_opportunity_dists_0.5.p", "rb") as file:
                    max_stats = pickle.load(file)
                    print(f"Max Stats = {max([max_stats["BVAP"][key] for key in max_stats["BVAP"].keys()])}")
```
